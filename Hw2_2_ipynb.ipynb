Частка пропусків у кожній колонці

```python
missing_values_ratio = df.isnull().sum() / len(df)
missing_values_ratio
```

Видаляємо усі стовпці з пропусками, крім "Мова програмування"

```python
columns_to_drop = missing_values_ratio[missing_values_ratio > 0].index.difference(['Мова програмування'])
df_dropped = df.drop(columns=columns_to_drop)
missing_values_ratio_new = df_dropped.isnull().sum() / len(df)
missing_values_ratio_new
```

Видалення рядків з пропусками

```python
df_clean = df_dropped.dropna()
```

Розмір таблиці після очищення

```python
df_clean.shape
```

Таблиця тільки з Python розробниками

```python
python_data = df_clean[df_clean['Мова програмування'] == 'Python']
```

Розмір таблиці Python розробників

```python
python_data.shape
```

Групування за стовпчиком "Посада" і агрегація даних за "Зарплата.в.місяць"

```python
grouped_data = python_data.groupby('Посада')
```

агрегація даних за "Зарплата.в.місяць"

```python
grouped_data = grouped_data['Зарплата на місяць'].agg(['min', 'max'])
grouped_data
```

Створення функції для обчислення середньої зарплати
Створення нового стовпця "avg" з середньою зарплатою

```python
def fill_avg_salary(row):
    return (row['min'] + row['max']) / 2

grouped_data['avg'] = grouped_data.apply(fill_avg_salary, axis=1)
grouped_data
```

Описова статистика для "avg"

```python
grouped_data['avg'].describe()
```
